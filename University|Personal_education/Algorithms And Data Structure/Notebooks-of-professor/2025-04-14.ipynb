```python
tree1 = {'root': {'P': None, 'L': 3, 'R': 8},
        3: {'P': 'root', 'L': 9, 'R': None},
        8: {'P': 'root', 'L': 2, 'R': 7},
        9: {'P': 3, 'L': None, 'R': None},
        2: {'P': 8, 'L': None, 'R': None},
        7: {'P': 8, 'L': None, 'R': None}}
```

```python
def add_root(newroot):
    newtree = dict()
    newtree[newroot] = {'P': None, 'L': None, 'R': None}
    return newtree
    
def add_child(tree, parent, side, newnode):
    # side: 'L' or 'R' (check if the side is already taken)
    if tree[parent][side] != None:
        print(parent, 'has already a ', side, 'child')
        return tree
    tree[parent][side] = newnode
    tree[newnode] = {'P': parent, 'L': None, 'R': None}
    return tree
```

### Implement the Python function <code>even_odd(xlist)</code> than inserts the values of a Python list into a binary tree so as to allocate the even numbers contained in the list on the left subtree, and the odd numbers on the right subtree. The order of insertion follows the order of appearance of the values in the original Python list.

```python
tree2
```

```python
def is_root(tree, node):
    return tree[node]['P'] == None

def parent(tree, node):
    return tree[node]['P']

def is_leaf(tree, node):
    return tree[node]['L'] == None and tree[node]['R'] == None

def children(tree, node):
    result = []
    if tree[node]['L'] != None:
        result.append(tree[node]['L'])
    if tree[node]['R'] != None:
        result.append(tree[node]['R'])
    return result
```

```python
is_root(tree2, 20)
```

```python
is_root(tree2, 30)
```

```python
is_root(tree2, 'root')
```

```python
print(parent(tree2, 20))
```

```python
print(parent(tree2, 'root'))
```

```python
def depth(tree, start):
    if is_root(tree, start):
        return 0
    else:
        return 1 + depth(tree, parent(tree, start))
```

```python
def height(tree, node):
    if is_leaf(tree, node):
        return 0
    else:
        return 1 + max(height(tree, c) for c in children(tree, node))
```
