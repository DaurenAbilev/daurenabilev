```python
def bfs(graph, start):
    # adjacency list
    queue = [start]
    visited = []
    while queue:
        current = queue.pop(0)
        if current not in visited:
            visited.append(current)
        neighbours = graph[current]
        for node in neighbours:
            if node not in visited:
                queue.append(node)
        print(current, neighbours, queue)
    return visited
        
        
```
