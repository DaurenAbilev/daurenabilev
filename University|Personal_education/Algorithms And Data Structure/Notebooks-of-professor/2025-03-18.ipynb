```python
def insertionsort(a):
    for j in range(1, len(a)):
        key = a[j]
        i = j
        while i > 0 and a[i-1] > key:
            a[i] = a[i-1]
            i = i - 1
            a[i] = key
            print(a)
    return a
```

```python
a = [5, 2, 4, 6, 1, 3]
```

```python
insertionsort(a)
```

```python
insertionsort([1, 2, 3, 4, 5, 6])
```
