## A vending machine returns only coins each worth 10 cents. A customer buys an item worth 20 cents using a 50 cents coin. How many coins he will have in exchange?

```python
def exchange(itemprice, insertedcoin, exchangecoin):
    residue = insertedcoin - itemprice
    return residue//exchangecoin
```

```python
exchange(20, 50, 10)
```

```python
exchange(20, 50, 5)
```

## A grading system has the following scores: A) 100-90, B) 89-80, C) 79-70, D) 69-60, F) 59-50. Ask the user to enter an integer score in the range 50-100 and print the corresponding literal score. For example, 73 corresponds to C.

```python
score = int(input("Enter a numeric score (50-100): "))
if score < 50 or score > 100:
    print("Out of range")
elif 50 <= score <= 59:
    print("F")
elif 60 <= score <= 69:
    print("D")
elif 70 <= score <= 79:
    print("C")
elif 80 <= score <= 89:
    print("B")
elif 90 <= score <= 100:
    print("A")
```

## Write the function between(x,y,z) which returns True if x ≤ y ≤ z, False otherwise

```python
def between(x, y, z):
    #c = x <= y <= z
    #return c
    #return x <= y <= z
    return x <= y and y <= z
```
