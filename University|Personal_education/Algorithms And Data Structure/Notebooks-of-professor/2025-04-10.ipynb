## Queues on doubly linked lists

```python
# insertion: through tail
# extraction: through head
def traverse(doublylinkedlist):
    current = 'head'
    while current != 'tail':
        next = doublylinkedlist[current]['next']
        print(current, '<==>', next)
        current = next
def insert_queue(doublylinkedlist, newnode):
    node1 = doublylinkedlist['tail']['prev']
    doublylinkedlist['tail']['prev'] = newnode
    doublylinkedlist[node1]['next'] = newnode
    doublylinkedlist[newnode] = {'prev': node1, 'next': 'tail'}
    
def extract_queue(doublylinkedlist):
    node1 = doublylinkedlist['head']['next']
    node2 = doublylinkedlist[node1]['next']
    doublylinkedlist['head']['next'] = node2
    doublylinkedlist[node2]['prev'] = 'head'
    del doublylinkedlist[node1]
    return node1
```

```python
dll_queue1 = {'head': {'prev': None, 'next': 'tail'},
             'tail': {'prev': 'head', 'next': None}}
xlist = ['a', 'b', 'c', 'd']
for x in xlist:
    insert_queue(dll_queue1, x)
traverse(dll_queue1)    
```

```python
# insertion through head
# extraction through tail

def insert_queue2(doublylinkedlist, newnode):
    node1 = doublylinkedlist['head']['next']
    doublylinkedlist[newnode] = {'prev': 'head', 'next': node1}
    doublylinkedlist['head']['next'] = newnode
    doublylinkedlist[node1]['prev'] = newnode
    
def extract_queue2(doublylinkedlist):
    node1 = doublylinkedlist['tail']['prev']
    node2 = doublylinkedlist[node1]['prev']
    doublylinkedlist['tail']['prev'] = node2
    doublylinkedlist[node2]['next'] = 'tail'
    del doublylinkedlist[node1]
    return node1
```

## Exchange the relative positions of two adjacent nodes in a doubly linked list

Before: left <==> node1 <==> node2 <==> right

After: left <==> node2 <==> node1 <==> right

```python
dll1 = {'head': {'prev': None, 'next': 'tail'},
             'tail': {'prev': 'head', 'next': None}}
```

```python
dll1['node1'] = {'prev': 'head', 'next': 'node2'}
dll1['node2'] = {'prev': 'node1', 'next': 'tail'}
dll1['head']['next'] = 'node1'
dll1['tail']['prev'] = 'node2'
```

```python
traverse(dll1)
```

```python
#Before: left <==> node1 <==> node2 <==> right

#After: left <==> node2 <==> node1 <==> right
    
def exchange_adj_nodes(doublylinkedlist, x1, x2):
    left = doublylinkedlist[x1]['prev']
    right = doublylinkedlist[x2]['next']
    doublylinkedlist[x1]['prev'] = x2
    doublylinkedlist[x1]['next'] = right
    doublylinkedlist[x2]['prev'] = left
    doublylinkedlist[x2]['next'] = x1
    doublylinkedlist[left]['next'] = x2
    doublylinkedlist[right]['prev'] = x1    
```

```python
exchange_adj_nodes(dll1, 'node1', 'node2')
```

```python
traverse(dll1)
```

def traverse_sum(doublylinkedlist):
    current = 'head'
    mysum = 0
    while current != 'tail':
        next = doublylinkedlist[current]['next']
        if current != 'head':
            mysum += 1
        #print(current, '<==>', next)
        current = next
    return mysum

```python
def insert(doublylinkedlist, prec, succ, newnode):
    if doublylinkedlist[prec]['next'] != succ or \
    doublylinkedlist[succ]['prev'] != prec:
        print(prec, succ, "not adjacent")
        return doublylinkedlist
    doublylinkedlist[newnode] = {'prev': prec, 'next': succ}
    doublylinkedlist[prec]['next'] = newnode
    doublylinkedlist[succ]['prev'] = newnode
    return doublylinkedlist
```

## Implement the Python function <code>dll_max(doublylinkedlist)</code> that return the highest number stored in the doubly linked list. Hint: traverse the doubly linked list.

```python
def dll_max(dll):
    xmax = dll['head']['next']
    current = dll['head']['next']
    while current != 'tail':
        if isinstance(current, (int, float)) and current > xmax:
            xmax = current
        current = dll[current]['next']
    return xmax
```
