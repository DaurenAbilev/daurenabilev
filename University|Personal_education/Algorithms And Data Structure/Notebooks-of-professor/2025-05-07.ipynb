```python
fin = open('soc-karate-edgelist.txt')
edgelist = []
for line in fin:
    #print(line)
    u, v = line.split()
    edgelist.append((int(u), int(v)))
    
for edge in edgelist:
    if not isinstance(edge[0], int) or not isinstance(edge[1], int):
        print(u, v, "Error")
```

```python
def bfs(graph, start):
    # adjacency list
    queue = [start]
    visited = []
    while queue:
        current = queue.pop(0)
        if current not in visited:
            visited.append(current)
        neighbours = graph[current]
        for node in neighbours:
            if node not in visited:
                queue.append(node)
        #print(current, neighbours, queue)
    return visited
```

```python
def el2al(graph):
    result = dict()
    for node1, node2 in graph:
        if node1 in result:
            result[node1].append(node2)
        else:
            result[node1] = [node2]
        if node2 in result:
            result[node2].append(node1)
        else:
            result[node2] = [node1]
    return result
```

```python
adjacencylist = el2al(edgelist)
```

```python
vertices = bfs(adjacencylist, 17)
```
