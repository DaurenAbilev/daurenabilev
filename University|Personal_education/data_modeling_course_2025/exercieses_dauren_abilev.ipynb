```python
filename = "esperimento_mru_or.txt"

with open(filename, "r", encoding="utf-8") as f:
    lines = f.read().strip().splitlines()

print(lines)
```

```python
headers = lines[0].strip().split(",")

data = {}
for h in headers:
    data[h] = []                        


for line in lines[1:]:
    t_str, s_str = line.strip().split(",")  
    data[headers[0]].append(int(t_str))
    data[headers[1]].append(int(s_str))   

print(data)
```
