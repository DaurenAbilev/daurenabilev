# Introduction to Data modeling

The essence of data modeling is creating a “digital twin” of real-world phenomena. Instead of running thousands of expensive or dangerous experiments in reality, we describe how an object behaves using mathematics.

This process consists of several simple steps:

1. Observation: we observe how something works in the real world.
2. Mathematics: we describe this behavior using formulas (for example, how fast a ball falls).
3. Programming: we translate these formulas into computer code, for example using Python.
4. Simulation: the computer “plays out” different scenarios, helping us predict outcomes or find solutions to problems.


This allows scientists to study everything—from the motion of planets to the development of new medicines—simply by running code on a laptop.

Simulation is a mathematical model translated into code that allows us to imitate the behavior of systems over time and under different conditions.

> Program is a sequence of operations designed for solving a specific class of problems

Algorithm is well defined caluculation procedure, which takes a set of values as input and produces a set of data as output

**Compiler** translates whole code into machine code and only after executes it, while **interpeter** execute code line by line
