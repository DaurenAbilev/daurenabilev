```python
from sklearn.datasets import load_iris
import pandas as pd
iris=load_iris()
iris['data']
```

```python
iris_df = pd.DataFrame (iris['data'], columns=['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width'])
iris_df.columns
iris_df['Sepal_Length']
```

```python
iris_df["Sepal_Length"]
```

```python
import numpy as np
np.mean(iris_df["Sepal_Length"]) #via mean value we can use it to replace empty value
```

```python
iris_df = pd.concat([iris_df ,pd.DataFrame (iris_df['target'], columns=['Species'])], axis=1)
```

```python
iris_df.groupby("Species"),["Sepal_Length"].mean()
```
