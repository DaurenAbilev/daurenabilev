# What is CRON? 

*CRON(utility) is task schedule that is used to starts any type of scipt on needed time. Used for automation*

- CRON is a program that is installed and run as a deamon 
- CRONJOB is a task that is needed to be completed in terms of CRON, any script or program 
- CRONTAB is a file that consist of the list of the tasks that are needed to be done and the time to be done at

---

( * * * * * ) - script/command

1st * - Minute(0-59) // also it is possible to use / to run some command every * mimute. */5 * * * * - stands for run every 5 minute

2nd * - Hour(0-23) // it is possible to use - to specify duration. f.e. run script from 5-8. 0 5-8 * * * - means it gonna run script from 5 to 8

3rd * - Day of the mounth(1-31) // also it is possbile to use , in case you need to specify 2 or more days. 0 0 1,15 * * - so it will run every 1st and 15th at midnight 

4th * - Mounth of the year(1-12)

5th * - Day of the week(0-6)
