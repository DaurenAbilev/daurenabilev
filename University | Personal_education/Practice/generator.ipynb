```python
def avg():
    for i in range(1,10):
        a = range(i, 11)
        yield sum(a) / len(a)
        
a = avg()
print(list(a))
```

```python
def find_word(f, word):
    g_index = 0 
    for line in f:
        indx = 0
        while (indx != -1):
            indx = line.find(word, indx)
            if indx > -1:
                yield g_index + indx
                indx += 1 
    g_index += len(line)
try: 
    with open("generator.txt", encoding="utf-8") as file:
        a = find_word(file, "generator")
        print(list(a))
except FileNotFoundError:
    print("File not Found")
except:
    print("Error")
```

```python
def even_numbers(number):
    for x in range(0, number, 2):
        yield x 

result = even_numbers(10)
print(list(result))
```

```python
def count_down(n):
    for x in range(n, 0, -1):
        yield x

a = count_down(5)

for i in a:
    print(i)
```

```python
def unique_in_order(data):
    y = None
    for x in data:
        if x != y:
            yield x 
            y = x

data = [1, 1, 2, 2, 2, 3, 1, 1, 4]

result = unique_in_order(data)
print(list(result))
```

```python
def chunk_size(item, size):
    answer = []
    for i in item: 
            answer.append(i)
            if len(answer) == size:
                yield answer
                answer = []
    if answer:
        yield answer            
        
data = [1, 2, 3, 4, 5, 6, 7]

result = chunk_size(data, 3)
print(list(result))
```

```python
def running_sum(data):
    summ = 0
    for i in data:
        summ+=i
        yield summ


data = [2, 5, -1, 10]

result = running_sum(data)
print(list(result))
```

```python
def sliding_window(items, size):
    answer = []
    for i in items:
        answer.append(i)
        if len(answer) == size:
            yield answer
            answer = answer[1:]

            

data = [1, 2, 3, 4, 5]
result = sliding_window(data, 3)
print(list(result))
```
