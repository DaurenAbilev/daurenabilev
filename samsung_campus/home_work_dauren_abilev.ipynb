```python
import pandas as pd

df = pd.read_csv('./auto-mpg.csv')

df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'name']
print(df.head(3))
print('\n'
```

miles -> km

galons -> l


## $\frac{(x) * miles}{galons} * \frac {1.609km}{miles} * \frac{1  galon}{3.785 l}$

### 0.425 differance
 

```python
df.dtypes
```

```python
df["mpg"] = df["mpg"].astype(float) # changed coloumn type for converting its values 
```

```python
df
```

```python
df["kml"] = (df["mpg"] * 0.425).round(1) # rounded kml coloum as it is formated for the table auto-mpg
```

```python
df
```

## $(L100km) = \frac{100}{kml}$

```python
df["L_100km"] = (100 / df["kml"]).round(1)
df
```

```python
df.sort_values("L_100km").head(5)[["name", "kml", "L_100km"]]
```
