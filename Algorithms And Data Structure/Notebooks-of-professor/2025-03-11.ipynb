### Implement the Python function <code>frequency(text)</code> that returns a dictionary containing the frequency of the letters in the argument string. For example, {'a': 3, 'o': 3, 'n': 4, ...}.

```python
def frequency(text):
    result = dict()
    for x in text:
        if x not in result:
            result[x] = 1
        else:
            result[x] = result[x] + 1
        print(x, result)
    return result
```

```python
frequency(string1)
```

### Define a dictionary that maps month name abbreviations to month names. For example, {'Jan': 'January', ...}. 1. create a list containing the 12 months, 2. create a dictionary whose keys are the month name abbreviations and the corresponding values are the full month names.

```python
months = ['January', 'February', 'March', 
          'April', 'May', 'June', 'July', 
          'August', 'September', 'October', 
          'November', 'December']
```

```python
abbreviations = dict()
for x in months:
    abbreviations[x[:3]] = x
print(abbreviations)
```

```python
def reverse_lookup(dictionary, value):
    result = []
    for k in dictionary:
        if dictionary[k] == value:
            result.append(k)
        print(k, dictionary[k], value, result)
    return result
```
