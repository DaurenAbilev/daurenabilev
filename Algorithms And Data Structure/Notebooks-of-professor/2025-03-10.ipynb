```python
def myreduce(a):
    total = 0
    for element in a:
        total = total + element
    return total
```

```python
myreduce([1,7, 4, 12, 24])
```

```python
def mymap(a, c):
    result = []
    for element in a:
        result.append(element * c)
    return result
```

```python
mymap([1, 7, 4, 12, 24], 2)
```

```python
def myfilter(a, c):
    result = []
    for element in a:
        if element%2 != 0:
            result.append(element * c)
    return result
```

```python
myfilter([1,7, 4, 12, 24], 2)
```

```python
def only_upper(s):
    result = []
    for x in s:
        if x.isupper():
            result.append(x)
    return result
```

```python
only_upper(['Hello', 'HOW', 'Are', 'YOU'])
```

## Implement a Python program that inputs some words from the user (using the <code>input()</code> function) and inserts them into a dictionary with the corresponding length.

```python
mydict = {}
while True:
    word = input("Enter a word ('stop' to end): ")
    if word == 'stop':
        break
    mydict[word] = len(word)
print(mydict)    
```
