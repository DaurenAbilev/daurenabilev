## Implement the Python function <code>nfind(stringx, charx)</code> that returns the number of occurrences of <code>charx</code> in <code>stringx</code>.

```python
def nfind(stringx, charx):
    counter = 0
    for i in range(len(stringx)):
        if stringx[i] == charx:
            counter += 1
    return counter
```

```python
nfind(mystring, 'x')
```

```python
nfind(mystring, 'e')
```

## Implement a Python program that prints each and every character of a string and checks if it is alphanumeric. Use first the C-like for loop and then the Pythonic for loop.

## Implement the Python function <code>index(stringx, charx)</code> that returns the position of the first occurrence of <code>charx</code> within the string <code>stringx</code>. If the character is not contained in the string, the function returns -1.

```python
def index(stringx, charx):
    for i in range(len(stringx)):
        if stringx[i] == charx:
            #print(i)
            return i
    return -1
```
