```python
def insert(queue, element):
    queue.append(element)
    return queue

def extract(queue):
    x = queue.pop(0)
    return x
    # return queue.pop(0)

def is_empty(queue):
    return len(queue) == 0

def top(queue):
    if not is_empty(queue):
        return queue[0]

def length(queue):
    return len(queue)
```
