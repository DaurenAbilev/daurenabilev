```python
def insert(doublylinkedlist, prec, succ, newnode):
    if doublylinkedlist[prec]['next'] != succ or \
    doublylinkedlist[succ]['prev'] != prec:
        print(prec, succ, "not adjacent")
        return doublylinkedlist
    doublylinkedlist[newnode] = {'prev': prec, 'next': succ}
    doublylinkedlist[prec]['next'] = newnode
    doublylinkedlist[succ]['prev'] = newnode
    return doublylinkedlist
```

```python
insert(dll, 'A', 'B', 10)
```

```python
def traverse(doublylinkedlist):
    current = 'head'
    while current != 'tail':
        next = doublylinkedlist[current]['next']
        print(current, '<==>', next)
        current = next
```

```python
traverse(dll)
```

## Implement the Python function <code>rtraverse(doublylinkedlist)</code> that traverses a doubly linked list starting from the tail node.

```python
def rtraverse(doublylinkedlist):
    current = 'tail'
    while current != 'head':
        prev = doublylinkedlist[current]['prev']
        #print(current, '<==>', prev)
        print(current, '<==>', end = ' ')
        current = prev
    print(current)
```

```python
rtraverse(dll)
```

```python
def sum_nodes(doublylinkedlist):
    current = 'head'
    s = 0
    while current != 'tail':
        if isinstance(current, (int, float)):
            s = s + current
        next = doublylinkedlist[current]['next']
        #print(current, '<==>', next)
        current = next    
    return s
```

```python
sum_nodes(dll)
```
