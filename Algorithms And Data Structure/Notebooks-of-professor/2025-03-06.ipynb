## Implement the Python function <code>mysum(listx)</code> that returns the sum of the values stored in the list. For example <code>mysum([11, 6, 8, 12, 16])</code> would return 53.

```python
def mysum(listx):
    total = 0
    for i in range(len(listx)):
        total = total + listx[i]
        print(i, listx[i], total)
    return total
```

## Implement the Python function <code>divide(listx, x)</code> that returns a list whose values are the elements of listx divided by x. For example <code>divide([11, 6, 8, 12, 16], 2)</code> would return [5.5, 3, 4, 6, 8].

```python
def divide(listx, x):
    for i in range(len(listx)):
        listx[i] = listx[i] / x
    return listx
```

```python
divide(list2, 2)
```

```python
divide([10, 20, 30, 40], 5)
```

```python
def mymax(listx):
    xmax = listx[0]
    for element in listx:
        if element > xmax:
            xmax = element
        print(element, xmax)
    return xmax
```

```python
mymax([6, 13, 9, 12, 7])
```

```python
def mymax2(listx):
    xmax = listx[0]
    imax = 0
    for i in range(len(listx)):
        if listx[i] > xmax:
            xmax = listx[i]
            imax = i
        print(i, "\t", listx[i], "\t", imax, "\t", xmax)
    return xmax, imax
```

```python
mymax2([6, 13, 9, 12, 27])
```

## Implement the Python function <code>incmatrix(listx, x)</code> that returns a list of lists whose values are incremented of the quantity x.

```python
def incmatrix(listx, x):
    for i in range(len(listx)):
        for j in range(len(listx[i])):
            listx[i][j] += x
            print(i, j, listx[i][j])
    return listx
```

```python
incmatrix(list5, 4)
```
