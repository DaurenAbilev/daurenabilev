```python
def add_root(newroot):
    newtree = dict()
    newtree[newroot] = {'P': None, 'L': None, 'R': None}
    return newtree
    
def add_child(tree, parent, side, newnode):
    # side: 'L' or 'R' (check if the side is already taken)
    if tree[parent][side] != None:
        print(parent, 'has already a ', side, 'child')
        return tree
    tree[parent][side] = newnode
    tree[newnode] = {'P': parent, 'L': None, 'R': None}
    return tree
```

```python
def left(tree, node):
    return tree[node]['L']

def right(tree, node):
    return tree[node]['R']

def sibling(tree, node):
    parent = tree[node]['P']
    if parent != None:
        if node == left(tree, parent):
            return right(tree, parent)
        else:
            return left(tree, parent)
    else:
        print(node, 'has no sibling node')
        return None
```

```python
print(left(tree1, 3))
print(left(tree1, 9))
print(right(tree1, 3))
print(right(tree1, 9))
print(sibling(tree1, 'root'))
print(sibling(tree1, 8))
print(sibling(tree1, 9))
print(sibling(tree1, 2))
print(sibling(tree1, 7))
```

```python
def preorder(tree, node, visited):
    if node != None:
        visited.append(node)
        lc = tree[node]['L']
        rc = tree[node]['R']
        if lc != None or rc != None:
            preorder(tree, lc, visited)
            preorder(tree, rc, visited)
        return visited
    
def preorder2(tree, node, visited):
    if node != None:
        visited.append(node)
        lc = tree[node]['L']
        rc = tree[node]['R']
        if lc != None or rc != None:
            preorder(tree, rc, visited)
            preorder(tree, lc, visited)
        return visited    
```

```python
def postorder(tree, node, visited):
    if node != None:
        lc = tree[node]['L']
        rc = tree[node]['R']
        if lc != None or rc != None:
            postorder(tree, lc, visited)
            postorder(tree, rc, visited)
        visited.append(node)      
        return visited
    
```
