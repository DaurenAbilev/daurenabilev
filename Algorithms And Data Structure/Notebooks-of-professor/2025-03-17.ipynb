## A proper subset is a special case of subset. The subset cannot be empty nor coincident to the superset.

```python
def proper_subset(set1, set2):
    # returns True if set1 is a proper subset of set2
    if set1 == set2:
        return False
    elif set1 == set():
        return False
    elif set1.issubset(set2):
        return True
    return False
```

```python
proper_subset(ca, it)
```

```python
proper_subset(ca, ca)
```

```python
proper_subset(ca, set2)
```

```python
proper_subset(it, fr)
```

```python
proper_subset(ca, fr)
```

```python
def proper_subset(set1, set2):
    if set1 == set():
        return False
    elif set1 == set2:
        return False
    for x in set1:
        if x not in set2:
            return False
    return True
```

## Implement the Python function <code>myunion(set1, set2)</code> which returns the union set of the two argument sets.

```python
def myunion(set1, set2):
    result = set()
    for x in set1:
        result.add(x)
    for x in set2:
        result.add(x)
    return result
```

```python
myunion(it, set())
```

## Implement the Python function <code>mydifference(set1, set2)</code> which returns the set difference of the two argument sets.

```python
def mydifference(set1, set2):
    """Returns the set containing the elements of set1 
    not belonging to set2
    """
    result = set()
    for x in set1:
        if x not in set2:
            result.add(x)
    return result
```

```python
mydifference(it, ca)
```
