```python
dll = {'head': {'prev': None, 'next': 100},
      100: {'prev': 'head', 'next': 150},
      150: {'prev': 100, 'next': 200},
      200: {'prev': 150, 'next': 'tail'},
      'tail': {'prev': 200, 'next': None}}
```

```python
# Remove node 150
dll[100]['next'] = 200
dll[200]['prev'] = 100
del dll[150]
```

```python
def insert(doublylinkedlist, prec, succ, newnode):
    if doublylinkedlist[prec]['next'] != succ or \
    doublylinkedlist[succ]['prev'] != prec:
        print(prec, succ, "not adjacent")
        return doublylinkedlist
    doublylinkedlist[newnode] = {'prev': prec, 'next': succ}
    doublylinkedlist[prec]['next'] = newnode
    doublylinkedlist[succ]['prev'] = newnode
    return doublylinkedlist

def traverse(doublylinkedlist):
    current = 'head'
    while current != 'tail':
        next = doublylinkedlist[current]['next']
        print(current, '<==>', next)
        current = next
```

```python
traverse(dll)
```

```python
def remove(dll, node):
    prec = dll[node]['prev']
    succ = dll[node]['next']
    dll[prec]['next'] = succ
    dll[succ]['prev'] = prec
    del dll[node]
```

```python
insert(dll, 100, 200, 150)
```

```python
remove(dll, 150)
```

```python
traverse(dll)
```

## Stack implemented on a doubly linked list

```python
# open end: tail

def insert_stack(doublylinkedlist, newnode):
    prec = doublylinkedlist['tail']['prev']
    doublylinkedlist[prec]['next'] = newnode
    doublylinkedlist['tail']['prev'] = newnode
    doublylinkedlist[newnode] = {'prev': prec, 'next': 'tail'}

def extract_stack(doublylinkedlist):
    node1 = doublylinkedlist['tail']['prev']
    if node1 != 'head':
        node2 = doublylinkedlist[node1]['prev']
        doublylinkedlist[node2]['next'] = 'tail'
        doublylinkedlist['tail']['prev'] = node2
        x = node1
        del doublylinkedlist[node1]
        return x
```

```python
dll_stack = {'head': {'prev': None, 'next': 'tail'},
       'tail': {'prev': 'head', 'next': None}}
```

```python
insert_stack(dll_stack, 'x1')
insert_stack(dll_stack, 'x2')
insert_stack(dll_stack, 'x3')
insert_stack(dll_stack, 'x4')
insert_stack(dll_stack, 'x5')
traverse(dll_stack)
```

```python
print(extract_stack(dll_stack))
traverse(dll_stack)
```

## Implement the <code>hinsert_stack()</code> and <code>hextract_stack</code> functions using the head of a doubly linked list as the only access end.

head <==> d <==> c <==> b <==> a <==> tail

```python
def hinsert_stack(doublylinkedlist, newnode):
    succ = doublylinkedlist['head']['next']
    doublylinkedlist[newnode] = {'prev': 'head', 'next': succ}
    doublylinkedlist['head']['next'] = newnode
    doublylinkedlist[succ]['prev'] = newnode

def hextract_stack(doublylinkedlist):
    node1 = doublylinkedlist['head']['next']
    node2 = doublylinkedlist[node1]['next']
    doublylinkedlist['head']['next'] = node2
    doublylinkedlist[node2]['prev'] = 'head'
    x = node1
    del doublylinkedlist[node1]
    return x
```

```python
dll_stack4 = {'head': {'prev': None, 'next': 'tail'},
       'tail': {'prev': 'head', 'next': None}}
hinsert_stack(dll_stack4, 'a')
hinsert_stack(dll_stack4, 'b')
hinsert_stack(dll_stack4, 'c')
hinsert_stack(dll_stack4, 'd')
traverse(dll_stack4)
```

```python
hextract_stack(dll_stack4)
```

```python
traverse(dll_stack4)
```
