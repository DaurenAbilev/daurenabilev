```python
import random
```

```python
data = []
while len(data) < 16:
    x = random.randint(1, 30)
    if x not in data:
        data.append(x)
data.sort()
print(data)
```

```python
def binarysearch(data, target, low, high):
    if low > high:
        return False
    else:
        mid = (low + high)//2
        print("low: ", low, "high: ", high, 'mid: ', mid)
    
    if target == data[mid]:
        print("Found")
        return True
    elif target < data[mid]:
        print("Left half")
        return binarysearch(data, target, low, mid-1)
    else:
        print("Right half")
        return binarysearch(data, target, mid+1, high)
```

```python
binarysearch(data, 21, 0, len(data))
```

## Stack

```python
def push(stack, element):
    stack.append(element)
    return stack
def pop(stack):
    x = stack.pop()
    return x
def top(stack):
    return stack[-1]
def is_empty(stack):
    return len(stack) == 0
def length(stack):
    return len(stack)
```

```python
stack1 = []
push(stack1, 'Alice')
push(stack1, 'Anitha')
print(stack1)
element = pop(stack1)
print("Extracted: ", element)
print(stack1)
print(top(stack1))
print(is_empty(stack1))
stack2 = []
print(is_empty(stack2))
print(length(stack1))
print(length(stack2))
```
