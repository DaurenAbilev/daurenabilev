```python
def vertex_count(graph):
  return len(graph[0])

def edge_count(graph):
  return len(graph) - 1

def vertices(graph):
  return graph[0]

def edges(graph):
  return graph[1:]

def get_edge(graph, u, v):
  return (u,v) in graph[1:] or (v, u) in graph[1:]

def degree(graph, v):
  count = 0
  for x in graph[1:]:
    if v in x:
      count += 1
  return count

def incident_edges(graph, u):
  result = []
  for edge in graph[1:]:
    if u in edge:
      result.append(edge)
  return result

def insert_vertex(graph, u):
  if u not in graph[0]:
    graph[0].append(u)

def remove_vertex(graph, u):
  if u in graph[0]:
    candidates = incident_edges(graph, u)
    for node1, node2 in candidates:
      remove_edge(graph, node1, node2)
    graph[0].remove(u)
  return graph

def insert_edge(graph, u, v):
  if u not in graph[0] or v not in graph[0]:
    print("At least one of the node does not exist")
    return graph
  if get_edge(graph, u, v):
    print("Edge already existing")
    return graph
  graph.append((u, v))
  return graph
  
def remove_edge(graph, u, v):
  if get_edge(graph, u, v):
    if (u, v) in graph[1:]:
      graph.remove((u, v))
    elif (v, u) in graph[1:]:
      graph.remove((v, u))
  return graph
```

```python
def is_complete(graph):
  n = len(graph[0])
  tm = n*(n-1)/2
  rm = len(graph) - 1
  return tm == rm
```

## Implement the Python function <code>degrees(graph)</code> that returns the degree of every node of a graph.

```python
def degrees(graph):
  result = {}
  for node in graph[0]:
    result[node] = degree(graph, node)
  return result
```

### Implement the Python function <code>all_incident_edges(graph)</code> that uses the function <code>incident_edges(graph, u)</code> to return <b>all</b> the incident edges of each node of the graph.

```python
def all_incident_edges(graph):
    result = {}
    for node in graph[0]:
        result[node] = incident_edges(graph, node)
    return result
```
