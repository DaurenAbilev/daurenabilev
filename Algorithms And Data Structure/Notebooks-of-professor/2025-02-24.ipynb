# Implement the Python function <code>circle(r)</code> that returns the area and the length of a circle of radius <code>r</code>.

$$ Area = \pi \cdot r^2 $$

$$Length = 2 \cdot r \cdot \pi $$

```python
def circle(r):
    a = math.pi * r * r
    p = 2 * math.pi * r
    return a, p
```

```python
circle(10)
```

```python
aa, bb = circle(10)
print(round(aa, 2))
print(round(bb, 2))
```
