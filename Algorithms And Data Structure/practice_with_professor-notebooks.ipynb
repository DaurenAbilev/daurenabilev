```python
# альтернативное решение 

def is_valid_email(s):
    if s.count("@") != 1:
        return False
    if s.startswith("@") or s.endswith("@"):
        return False
    at = s.index("@")
    domain = s[at+1:]          # часть после @
    if "." not in domain:
        return False
    if domain.startswith(".") or domain.endswith("."):
        return False
    return True
```

```python
def char_types(text):
    types = {
                "letters": 0,
                "digits": 0,
                "spaces": 0
            }
    for i in range(len(text)):
        if text[i].isalpha():
            types["letters"]+=1
        elif text[i].isdigit():
            types["digits"]+=1
        elif text[i].isspace():
            types["spaces"]+=1
        else:
            pass
    print(types)


test = "dauren how old 123"
char_types(test)
```

Write the Python function onlyodd(xlist), xlist being a list of 10 integers in the range (1,20),
that returns a list containing only the odd numbers contained in the list xlist.

```python
def onlyodd(xlist):
    return [x for x in xlist if x%2 != 0] 

onlyodd([1,2,3,4,5,6])
```

Write a function digit_sum(n) that recursively returns the sum of the digits of an integer n.

```python
def digit_sum(n):
    if n == 0:
        return 0
    else:
        return n%10 + digit_sum(n//10)

digit_sum(3)
```

```python
def count_letters(text):
    d = {}
    for key in text:
        if key.isalpha():        # считаем только буквы
            if key in d:
                d[key] += 1
            else:
                d[key] = 1
    return d


print(count_letters("hello"))
```

```python
def group_by_length(words):

    a = {}

    for key in words:
        if len(key) in a:
            a[len(key)].append(key)
        else: 
            a[len(key)] = [key]
    return a

group_by_length(["a", "hi", "cat", "go", "dog"])
```

```python
def ds_max(lst):
    max_value = 0
    if len(lst) == 0:
        return None
    for element in lst:
        if element > max_value:
            max_value = element
    return max_value
    


    
lst = [1,2,3,4,456]
ds_max(lst)
```

```python
def duplicates(lst):
    seen = []
    duplicate = []
    for value in lst:
        if value in seen:
            duplicate.append(value)
        else:
            seen.append(value)
    return duplicate



lst = [1,1,2,3,3,4]
duplicates(lst)
```

```python
def unique(lst):
    unique = []
    for i in lst:
        if i not in unique:
            unique.append(i)
    return unique

unique([1,2,2,3,3,5,2])
```

```python
def pangram(text):
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    text = text.lower()
    for ch in alphabet:
        if ch not in text:
            return False
    return True

```

```python
def common_elements(a, b):
    unique = []
    for element in a:
        if element in b:
            if element not in unique:
                unique.append(element)
    return unique
```
