## Insertion Sort

Insertion Sort builds the final sorted array one element at a time by inserting each new element into its correct position within the already sorted part (like cards)

---

Algorithm Type
• Iterative
• In-place (no extra array needed)
• Comparison-based sorting

---

When to Use
• Small datasets
• Nearly sorted arrays
• As part of hybrid algorithms

Key Property
Insertion Sort is stable (equal elements keep their relative order).


```python
# The computational cost of an algorithm depends on the input size

def InsertionSort(a):
    for j in range(1, len(a)):
        key = a[j]
        i = j
        while i > 0 and a[i-1] > key:
            a[i] = a[i-1]
            i = i - 1
            a[i] = key
        print(a)
    return a

list = [4,2,5,1,9]

print(InsertionSort(list))
```
