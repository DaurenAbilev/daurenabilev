## Recursion 

- Python functions can invoke themselves. It discribes recursion 

If a recursive function <u>never reaches the base case</u>, the
function call will be executed an infinite number of times it called <u>infinite recursion</u>


```python
def countdown(n):
    if n <= 0:
        print("Go!!!")
    else:
        print(n)
        countdown(n-1)

countdown(5)
```

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

print(factorial(6))
```

```python
# A number a is power of b if it is divisible by b and if a/b is a power of b. Write a function named power(a,b) which
# requires two arguments (a and b) and returns True if a is a power of b.

def power(a, b:int) -> int:
    if a == 1:
        return True
    
    if a % b == 0 and a != 0:
        return power(a // b, b)
    
    return False

power(16,2)
```

```python
# Write the recursive function product(n) which requires an integer number n as argument and returns its product with all
# natural numbers that precede it. For example, product(4) must return 24= 4· 3· 2· 1

def product(n:int) -> int:
    if n <= 1:
        return 1
    else: return n * product(n-1)

product(4)
```

```python
# Searching

def find(word, letter):
    index = 0
    while index < len(word):
        if word[index] == letter:
            return index + 1
        index = index + 1
    return -1

find("text", "x")
```

– Reduce: calculate the sum of the elements of a numeric list

– Map: apply the same operation to all elements of a list

– Filter: apply some operation to some elements of a list

```python
# Implement the Python function myreduce(a) that returns the sum of the elements of the list a.

def myreduce(a):
    summing = 0
    for element in range(len(a)):
        summing = a[element] + summing
    return summing

def myreduce_upd(a):
    summing = 0
    for element in a:
        summing += element
    return summing

a = [1,2,3,4,5,6]

myreduce(a)
```

```python
# Implement the Python function mymap(a) that returns a list containing the elements of the list a each multiplied by 2.

def mymap(a):
    new_list = []
    for element in range(len(a)):
        new_list.append(a[element] * 2)
    return new_list

def mymap_upd(a):
    new_list = []
    for element in a:
        new_list.append(element * 2)
    return new_list
a = [1,2,3,4,5,6]

mymap_upd(a)
```

```python
# Implement the Python function myfilter(a) that returns a list containing the odd elements of the list a each multiplied by 2

def myfilter(a):
    new_list = []
    for i in a:
        if i%2 != 0:
            new_list.append(i * 2)
    return new_list

myfilter(a)
```

### Aliasing

• Aliasing: This occurs when a single object in memory has more than one name or reference.

• Mechanism: The assignment b = a does not create a copy of the object; it simply associates the name b with the same physical object already referenced by a.

• Mutability: Because both variables point to the same address, any change made to a mutable object (like a list) through one name is immediately reflected in all of its aliases.

• Identification: You can confirm if two variables are aliases by using the id() function, which returns a unique identifier for the object's location in memory; aliases will always have the same id.

---

### is 

• == (Equivalence): Use this to check if two objects contain the same values or elements.

• is (Identity): Use this to check if two variables refer to the exact same object in memory. Two strings or lists might be equivalent (== is True) but not identical (is is False) if they are stored in different memory locations.

Common Practical Use
• In real-world algorithms, is is most frequently used to check if a variable is None. This is a standard way to verify if a node in a tree or a linked list exists before attempting to access its data.

## Dictionary

– A dictionary represents the generalization of a list

– In a list indices are integer numbers, whereas in a dictionary indices can be (almost) every kind of datum

– A dictionary is composed of a collection of indices (keys) and a collection of values

– To every key corresponds one and only one value

– The elements of a dictionary are the pairs keys-value

```python
phonebook = {"joe": "090123", "frank": "090456", "ann": "090789"}

print(phonebook["ann"])
print(phonebook.get("frank"))

phonebook.get("Dauren", 0) # A "safer" way to lookup a value. It returns the value for the key if it exists; otherwise, it returns the provided default value (prevents errors).
phonebook.keys() # Returns a collection of all keys in the dictionary.
phonebook.values() # Returns a collection of all values in the dictionary.
phonebook.items() # Returns a sequence of tuples, where each tuple is a key-value pair.
```

Mutability: Dictionaries are mutable; you can add, remove, or update pairs at any time.

Efficiency: They are "very good" for tasks like counting frequencies (e.g., counting letters in a string) because they handle dynamic keys efficiently.

Complex Values: A value in a dictionary can be a list or another dictionary, which is how linked lists and trees are implemented in your course.

```python
# Count how many times each letter is contained in a given string

def count_letters(text):
    dictionary = {}
    for c in text:
        if c in dictionary:
            dictionary[c] += 1
        else:
            dictionary[c] = 1
    return dictionary

dic = count_letters("i am just checking")

print(dic.get("c", 0))
```

```python
# Reverse lookup: given a dictionary and a value, find the corresponding key

def reverse_lookup(dic, val):
    for k in dic:
        if dic[k] == val:
            return k
        print("Not found")
```

```python
def invert_dic(dic):
    inverse = dict()
    for k in dic:
        val = dic[k]
        if val not in inverse:
            inverse[val] = [k]
        else:
            inverse[val].append(k)
    return inverse

dic = {"a": 1, "p": 2, "l": 1, "l": 1}
invert_dic(dic)
```

```python
#Write the Python function mysubset(s,t) that returns True if set s is a subset of set t. Do not use the set function issubset().

def mysubset(s, t):
    for element in s:
        if element not in t:
            return False
    return True
```

```python
# Write the Python function mypropersubset(s,t) that returns True if set s is a proper subset of set t. A set is
# proper subset of another set if it is a subset, but the sets are not equal and the subset is not empty. Do not use the set function issubset().

def mypropersubset(s, t):
    
    if not s:
        return False
    if s == t:
        return False
    for element in s:
        if element not in t:
            return False
    
    return True
```

The <u>union</u> of two sets contains all of the elements of both sets, with duplicates removed

A new set is returned (neither of the sets is modified)

```python
ca = {"red", "white"}
it = {"red", "white", "green"}

ca_u_it = ca.union(it) # ca_u_it = ca|it

print(ca_u_it)
```

```python
# Write the Python function myunion(s,t) that returns the set union of the sets s and t. Do not use the set function union().

def myunion(s, t):
    new_set = set()
    
    for element in s:
        new_set.add(element)
    for element in t:
        new_set.add(element)
        
    return new_set    
```

```python
# Write the Python function myintersection(s,t) that returns the set intersection of the sets s and t. Do not use the set function intersection().

def myintersection(s, t):
    new_set = set()
    for element in s:
        if element in t:
            new_set.add(element)
    return new_set
```

```python
# Write the Python function frequency(text) that counts how often each word occurs in a text file.

def frequency(text):
    freq = {}
    words = text.split()
    
    for word in words:
        word = word.lower()          
        if word in freq:
            freq[word] += 1
        else:
            freq[word] = 1
            
    return freq
```
